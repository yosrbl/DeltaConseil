```python
options = webdriver.ChromeOptions()


#driver = webdriver.Chrome('C:/Users/amalh/Downloads/chromedriver.exe')

df = pd.DataFrame()

for i in range(20,180,20):
    
    driver.get("https://www.indeed.fr/cmp/Macdonnald's/reviews?fcountry=FR&sort=helpfulness?start="+str(i))
    
    driver.implicitly_wait(4)
    for job in driver.find_elements_by_class_name('cmp-Review'):
        #x=len(driver.find_elements_by_class_name('cmp-Review'))
        soup = BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')
        #print(i)
        try:
            review_body = soup.find("span",class_="cmp-NewLineToBr-text").text.replace("\n","").strip()
            #print(review_body)
        except:
            review_body = 'none'
        try:
            review_title = soup.find("div",class_="cmp-Review-title").text.replace("\n","").strip()
            #print(review_title)
            
        except:
            review_title = 'none'
        try:
            bubble_rating = soup.find("div",class_="cmp-ReviewRating-text").text.replace("\n","")
            #print( bubble_rating)
        except:
            bubble_rating = 'none' 
        try:
            a = soup.find("span",class_="cmp-ReviewAuthor").text.replace(" - ","ff").split("ff")
            #print(a)
            poste=a[0]
            location=a[1].split("(")[0]
            date=a[2]
            num_reviews=(a[1].split("(")[1]).split(")")[0]
        except:
            location= 'none' 
            poste= 'none'
            date= 'none' 
            num_reviews= 'none' 

        df = df.append({'review_title':review_title,"review_body":review_body,"bubble_rating":bubble_rating
                        ,"location":location,"poste":poste,'date':date,'num_reviews':num_reviews},ignore_index=True) 
                        
        print("Got these many results:",df.shape)
df.to_csv("mcdo.csv",index=False)
```
